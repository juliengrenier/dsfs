# Bar Charts

```python
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Ganhdi", "West Side Story"]
num_oscars = [5,11,3,8,10]
plt.bar(range(len(movies)), num_oscars)
plt.title('Movies')
plt.ylabel("# of oscars")
plt.xticks(range(len(movies)), movies)
plt.show()
```

```python
grades = [93, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]
histogram = Counter(min(grade // 10 * 10, 90) for grade in grades)
plt.bar([x + 5 for x in histogram.keys()], # Shift to the center
        histogram.values(), # Number of students
        10, # Width of 10
        edgecolor='black'
       )
plt.axis([-5, 105, 0, 5])
plt.xticks([10 * i for i in range(11)]) # X labels
plt.xlabel("Decile")
plt.ylabel("# of students")
plt.title("Distribution of grades")
plt.show()
```

```python
mentions = [500, 505]
years = [2017, 2018]
plt.bar(years, mentions, 0.8)
plt.xticks(years)
plt.ylabel("# of times hearing Data Science")
plt.ticklabel_format(useOffset=False)
plt.axis([2016.5, 2018.5, 499, 506])
plt.title("Misleading")
plt.show()
```

```python
plt.bar(years, mentions, 0.8)
plt.xticks(years)
plt.ylabel("# of times hearing Data Science")
plt.ticklabel_format(useOffset=False)
plt.axis([2016.5, 2018.5, 0, 550])
plt.title("Not Misleading")
plt.show()
```

# Line Chart

```python
variances = [2**i for i in range(9)]
bias_squared = [2**i for i in range(8, -1, -1)]
total_error = [x+y for x,y in zip(variances, bias_squared)]
xs = [i for i,_ in enumerate(variances)]
```

```python
plt.plot(xs, variances, 'g-', label='variance') # green solid
plt.plot(xs, bias_squared, 'r-.', label='bias^2') # red dashed
plt.plot(xs, total_error, 'b:', label='total error') #blue dotted
plt.legend(loc=9)
plt.xlabel('model complexity')
plt.xticks([])
plt.title('The bias-variance tradeoff')
plt.show()
```

# Scatterplots

```python
friends = [70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']
plt.scatter(friends, minutes)
for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(
        label,
        xy=(friend_count, minute_count),
        xytext=(5, -5),
        textcoords='offset points'
    )
plt.title('Daily minutes vs number of friends')
plt.xlabel('# of friends')
plt.ylabel('daily minutes spent on site')
plt.show()
```
